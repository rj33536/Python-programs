```python
import numpy as np
import matplotlib.pyplot as plt
class LinearRegressor:
    def __init__(self,iterations):
        self.weight=[]
        self.learning_rate=0.01
        self.iterations=iterations
        
    def fit(self,trainX,trainY):
        self.weight=np.ones(trainX.shape[1]+1)
        bias=np.ones((trainX.shape[0],1))
        trainX=np.concatenate((trainX,bias),axis=1)
        for i in range(self.iterations):
            output=np.matmul(trainX,self.weight)
            gradient=np.matmul(trainX.T,output-trainY)
            print(gradient)
            self.weight-=self.learning_rate*gradient
            
    def predict(self,inp):
        return (np.matmul(np.concatenate((inp,np.ones((inp.shape[0],1))),axis=1),self.weight))
```

```python
s=LinearRegressor(1000)
x=np.array([[1,1],[0,1]])
y=np.transpose(np.array([2,3]))
s.fit(x,y)
print(s.predict(np.array([[0,1],[1,1]])))
```

```python
import numpy as np

class LogisticRegression:
    def __init__(self,iterations):
        self.weight=[]
        self.learning_rate=0.01
        self.iterations=iterations
        
    def fit(self,trainX,trainY):
        self.weight=np.ones(trainX.shape[1]+1)
        bias=np.ones((trainX.shape[0],1))
        trainX=np.concatenate((trainX,bias),axis=1)
        for i in range(self.iterations):
            output=(np.matmul(trainX,self.weight))
            output/=abs(output)
            gradient=np.matmul(np.transpose(trainX),output-trainY)
            self.weight=self.weight-self.learning_rate*gradient
            
    def predict(self,inp):
        output = np.matmul(np.concatenate((inp.T,[1])),self.weight)
        return output/abs(output)
```

```python
s=LogisticRegression(1000)
x=np.array([[1,0],[0,1]])
y=np.transpose(np.array([-1,1]))
s.fit(x,y)
a = []
b = []
for i in range(5):
    for j in range(5):
        prediction=s.predict(np.array([i,j]))
        if(prediction==1):
            a.append(prediction)
        else:
            b.append(prediction)

fig,ax = plt.subplots()
ax.scatter(a[:2],b[:2])
```

```python
import numpy as np

class svm:
    def __init__(self,margin,iterations):
        self.weight=[]
        self.margin=margin
        self.learning_rate=0.01
        self.iterations=iterations
        
    def fit(self,trainX,trainY):
        self.weight=np.ones(trainX.shape[1]+1)
        bias=np.ones((trainX.shape[0],1))
        trainX=np.concatenate((trainX,bias),axis=1)
        for i in range(self.iterations):
            output=np.matmul(trainX,self.weight)
            gradient=np.matmul(np.transpose(trainX),output-trainY)
            self.weight=self.weight-self.learning_rate*gradient
            
    def predict(self,inp):
        distance = np.matmul(np.concatenate((inp.T,[1])),self.weight)/(sum(self.weight**2)-self.weight[0]**2)
        if distance<0:
            if distance<-1*self.margin:
                return -1
            else:
                return 0
        else:
            if distance>self.margin:
                return 1
            else:
                return 0
            
            
```

```python
s=svm(1,1000)
x=np.array([[1,1,0],[1,0,1]])
y=np.transpose(np.array([-1,1]))
s.fit(x,y)
print(s.predict(np.array([1,-2,2])))
```
